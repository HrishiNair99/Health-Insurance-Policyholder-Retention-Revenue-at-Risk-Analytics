```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('train_enriched.csv')
df.head(5)
```

```python
#counting no of empty cells in each column
df.isnull().sum()
```

```python
#counting % of empty cells in each column
df.isnull().mean()*100
# this indicated that around 62% people were never contacted hence thet have no campaign_type or campaign_response
#so leaving these blanks as it is
```

```python
#checking for duplicates in id column
dupli=df.duplicated(subset='id').sum()
print(dupli)
df.shape
```

```python
#removing duplicates
dfnd=df.drop_duplicates(subset='id')
```

```python
#checking for duplicates in id column again
dupli=dfnd.duplicated(subset='id').sum()
print(dupli)
dfnd.shape
```

```python
#Checking values of the categorical columns
print('Gender', dfnd['Gender'].unique())
print('Vehicle_Age', dfnd['Vehicle_Age'].unique())
print('Vehicle_Damage', dfnd['Vehicle_Damage'].unique())
print('campaign_type', dfnd['campaign_type'].unique())
print('lapse_risk_score', dfnd['lapse_risk_score'].unique())
```

```python
# Annual_Premium outlier fix
Q1 = dfnd['Annual_Premium'].quantile(0.25)
Q3 = dfnd['Annual_Premium'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Count outliers
outliers = ((dfnd['Annual_Premium'] < lower) | (dfnd['Annual_Premium'] > upper)).sum()
print(f'Annual_Premium outliers: {outliers}')

# Print min/max before capping
print(f'Before — min: {dfnd["Annual_Premium"].min()}, max: {dfnd["Annual_Premium"].max()}')

# Cap (winsorise) --- .clip()-- if value lower than lower then replace it with lower similarly for upper
dfnd['Annual_Premium'] = dfnd['Annual_Premium'].clip(lower, upper)

# Print min/max after capping
print(f'After — min: {dfnd["Annual_Premium"].min()}, max: {dfnd["Annual_Premium"].max()}')
```
